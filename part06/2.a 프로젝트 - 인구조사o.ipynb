## 실습 1. '삼청동' 의 인구 구조를,  0세부터 100세 까지 나이대 별로 몇명이 있는지 시각화 하세요. 
- 가로축은 나이, 세로축은 인구수

```python
import csv

data_list = []

f = open('age.csv', encoding='cp949')
data = csv.reader(f)
next(data)

for row in data :
    if '신림동' in row[0] :
        for i in row[3:]:
            i = i.replace(',', '')
            i = int(i)
            data_list.append(i)
            
print(data_list)
```

```python
len(data_list)
```

```python
x = np.arange(0, 100 + 1)
```

```python
y = data_list
```

```python
plt.plot(x, y )
```
